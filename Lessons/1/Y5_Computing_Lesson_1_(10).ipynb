### Writing and reading comments and Document strings  
> Comments and document strings (or docstrings) are programmer remarks and pertinent info meant for fellow programmers and users, and even oneself to read. They can also be used for debugging faulty code. Although docstrings main feature is in support of code documentation, it is particularly useful in helping to test code in chunks.  

> Add one or more ```#``` to the start of a phrase to make it a comment:

```
####################
#even index number#
####################
for i in range(1,29):   #loop for every student index number 1 to 28
    if i%2 == 0:        #checks if index is even number
        print(f'{i} is even index number')
```
> Docstrings are wrapped with triple inverted commas like this:
```'''this is a document string'''```

>Docstrings can be accessed via the ```help()``` function. THIS CAN BE USEFUL IN A PRACTICAL EXAM, and checking instructions on how to use a function is allowed during the exam.  
>Try accessing docstrings using ```help(print)``` for built-in functions or ```help(str.strip)``` for class based methods.

````python
help(print)
````

````python
help(str.strip)
````

> Example 4  

> Translate the pseudocode for validation of ```subjectcode``` below to python code, adding an input line and an output line for ```validFlag``` at the start and end of the program respectively.  

> ```subjectcode``` is used to uniquely identify a particular subject and is four digits. A valid ```subjectcode``` is 4 characters long and contain only digits.

image.png

````python
def validate_subject_code():

    subjectCode = input("Enter the subject code: ")

    validFlag = True

    if len(subjectCode) != 4:
        print("Error! Subject Code must be 4 digits.")
        validFlag = False

    elif not subjectCode.isdigit():
        print("Error! Subject Code contains only digits only.")
        validFlag = False

    print("validFlag =", validFlag)

validate_subject_code()
````
