### 1.1 Variable Increment  
Very often, programmers use variables to keep track of iterations (loops), counts, etc

What does this expression look like from a non-programmer's view?
```x = x + 1```

This expression is not a mathematical one, but rather a programming one. The "=" is not a check for equivalence, but rather it assigns the value on the right to the variable on the left of the term.

For example, ```x = 1``` means the value of 1 is assigned to the variable called x. In python, the variable is automatically created when its first reference, so there is no need to delcare the avariable or data type unlike in compiled languages like C.

Then why have two copies of x? The variable on the right represents the previous value the variable is holding. As such, ```x = x + 1```
is read as
``add 1 to the existing value of x and re-assign this new value to x```

The python shortcut for ```x=x+1``` is ```x+=1```

The same goes for these other decrements and other math operations like //, % :
```
x-=1 #x=x-1
x*=1 #x=x*1
x//=1 #x=x//1
x%=1 #x=x%1
```

````python
#try out the python increment shortcuts here
x=2
print(x)

x/1
print(x)

x//=1
print(x)

x*=2
print(x)

x-=2
print(x)

x%=2
print(x)
````

# 3. Logical Operators

Logical operators are used in programming to combine and manipulate boolean values (**True** or **False**). There are three common logical operators: **and, or**, and **not**. These operators allow you to perform logical operations on one or more expressions and produce a result based on the True/False values of those expressions.

These logical operators are widely used in programming to control the flow of execution based on certain conditions. They help in making decisions, combining multiple conditions, and performing complex logical operations. Typically, the conditions made with logical operators exist in a selection statement shown below, but we'll leave this topic to another lesson:  
```
if user_guess == answer and num_of_tries <=5:
    print('you win!')
else:
    print('sorry, try again')
```

````python
#Comment True or False at the end of each individual line before running the code.
#The first one has been done for you

print(35 < 32)  # False
print(32 > 8) # True
print(0 == (8 - 8)) # True

print("12 >= 12") # 12 >= 12
print(12 >= 12) # True
print("15 == None") # 15 == None
print(15 == None) # False

print("\n") # New Line
print("True and True, True or True")
print(3 > 1 and 3 > 2) # True
print(3 > 1 or 3 > 2) # True
print("False and False, False or False")
print(3 < 1 and 3 < 2) # False
print(3 < 1 or 3 < 2) # False
print("True and False, True or False")
print(3 > 1 and 3 < 2) # False
print(3 > 1 or 3 < 2) # True
````
